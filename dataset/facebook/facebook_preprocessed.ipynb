```python
# load the facebook network edges
edges_file = "dataset/facebook/facebook_combined.txt"
edges = pd.read_csv(edges_file, sep=" ", header=None, names=["source", "target"])

# check network structure
print(edges.head())
print(f"Total edges: {edges.shape[0]}")
print(f"Total unique users: {len(set(edges['source']).union(set(edges['target'])))}")
```

```python
# check basic statistics
num_edges = edges.shape[0]
unique_nodes = set(edges["source"]).union(set(edges["target"]))

print(f"Total edges: {num_edges}")
print(f"Total unique nodes: {len(unique_nodes)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# compute degree for each node
degree_counts = edges["source"].value_counts() + edges["target"].value_counts()

# plot degree distribution
plt.figure(figsize=(8, 5))
sns.histplot(degree_counts, bins=30, kde=True)
plt.yscale("log")  
plt.xlabel("Degree (Number of Connections)")
plt.ylabel("Frequency (Log Scale)")
plt.title("Degree Distribution of Facebook Network")
plt.show()
```

```python
all_nodes_in_edges = set(edges["source"]).union(set(edges["target"]))
isolated_nodes = set(range(max(all_nodes_in_edges))) - all_nodes_in_edges  # assuming node IDs are sequential

print(f"Total isolated nodes (not appearing in any edge): {len(isolated_nodes)}")
```

```python
# save edges dataset as final preprocessed network file
edges.to_csv("dataset/facebook/facebook_preprocessed.csv", index=False)

print("Export completed: dataset/Facebook/facebook_final_preprocessed.csv")
```
